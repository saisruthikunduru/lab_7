```python
import pandas as pd
data1=pd.read_csv('/content/breast_cancer_survival (1).csv')
data1
print(data1.head())
print(data1.columns)
target_variable = 'Patient_Status'
features = data1.columns[data1.columns != target_variable]
print('Target Variable:', target_variable)
print('Features:',features)
```

```python
data=data1.fillna(0)
```

```python
y=data['Patient_Status']
y
```

```python
y = y.replace({'Alive':1,'Dead':0})
y
```

```python
x=data.drop('Patient_Status',axis=1)
x
```

```python
x['Gender'] = x['Gender'].replace({'FEMALE': 1,'MALE':0})
x
```

```python
x['ER status'] = x['ER status'].replace({'Positive': 1, 'Negative' : 0})
x['HER2 status'] = x['HER2 status'].replace({'Positive': 1, 'Negative' : 0})
x['PR status'] = x['PR status'].replace({'Positive': 1, 'Negative' : 0})
x
```

```python
d4=x.drop('Histology',axis=1)
d1=d4.drop('Tumour_Stage',axis=1)
d2=d1.drop('Surgery_type',axis=1)
d3=d2.drop('Date_of_Surgery',axis=1)
d=d3.drop('Date_of_Last_Visit',axis=1)
d
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
d_train, d_test, y_train, y_test = train_test_split(d, y, test_size=0.3, random_state=42)
```

```python
k_values = list(range(2, 21))
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(d_train, y_train)
    accuracy = knn.score(d_test, y_test)
    accuracies.append(accuracy)
```

```python
print(accuracies)
```

```python
plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.title('Accuracy vs. K Value for KNN')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.show()
```

```python
#for k values 11 and 19 have highest accuracy
```

```python
import pandas as pd
c=pd.read_csv("/content/Japan_life_expectancy (1).csv")
print(c)
```

```python
y=c['Life_expectancy']
print(y)
```

```python
c.drop(labels='Prefecture',axis=1,inplace=True)
print(c)
```

```python
X=c.drop('Life_expectancy',axis=1)
X
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
d_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
k_values = list(range(1, 21))
mse_values = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)
print(mse_values)
```

```python
plt.plot(k_values, mse_values, marker='o', linestyle='-')
plt.title('Mean Squared Error vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
#for k value 1 the mean square error is maximum
#for k value 14 the mean square error is  minimum
```
